```python
import numpy as np
import pandas as pd
```

```python
import plotly as py
import plotly.graph_objs as go

from plotly.offline import download_plotlyjs, init_notebook_mode, iplot,plot
init_notebook_mode(connected=True)

import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

# LineCharts

```python
trace1=go.Scatter(x=df.index,
                 y=df.total_bill,
                 mode="lines",
                 name="total_bill",
                 text=df.total_bill)

data=[trace1]
fig=dict(data=data)
iplot(fig)
```

```python
trace1=go.Scatter(x=df.index,
                 y=df.tip,
                 mode="lines+markers",
                 name="total_bill",
                 text=df.total_bill)

data=[trace1]
fig=dict(data=data)
iplot(fig)
```

```python
trace1=go.Scatter(x=df.index,
                 y=df.total_bill,
                 mode="lines",
                 name="Total_Bill",
                 text="Total_Bill",
                 marker=dict(color="blue"))

trace2=go.Scatter(x=df.index,
                 y=df.tip,
                 mode="lines+markers",
                 name="Tip",
                 text="Tip",
                 marker=dict(color="purple"))

data=[trace1,trace2]
fig=dict(data=data)
iplot(fig)
```

# Scatter

```python
dfsat=df[df.day=="Sat"].copy()
dfsun=df[df.day=="Sun"].copy()
dfthur=df[df.day=="Thur"].copy()
dffri=df[df.day=="Fri"].copy()
```

```python
trace1=go.Scatter(x=dfsat.index,
                 y=dfsat.total_bill,
                 mode="markers",
                  name="Saturday")

trace2=go.Scatter(x=dfsun.index,
                 y=dfsun.total_bill,
                 mode="markers",
                 name="Sunday",
                 text="Sunday")

trace3=go.Scatter(x=dfthur.index,
                 y=dfthur.total_bill,
                 mode="markers",
                 name="Thursday",
                 text="Thursday")

trace4=go.Scatter(x=dffri.index,
                 y=dffri.total_bill,
                 mode="markers",
                 name="Friday",
                 text="Friday")



data=[trace1,trace2,trace3,trace4]
fig=dict(data=data)
iplot(fig)
```

# BarChar

```python
import plotly.io as pio
pio.templates.default="none"
```

```python
trace1=go.Bar(x=df.sex,
             y=df.total_bill,
             name="Total_Bill",
             marker = dict(color = 'rgba(255, 174, 255, 1)'))


trace2=go.Bar(x=df.sex,
             y=df.tip,
             name="Tip",
             marker = dict(color = 'rgba(255, 255, 128, 1)'))


data=[trace1,trace2]
layout = go.Layout(barmode = "group")
fig=dict(data=data,layout=layout)
iplot(fig)
```

```python
trace1=go.Bar(x=df.day,
             y=df.total_bill,
             name="total_bill",
             marker=dict(color="pink"))

trace2=go.Bar(x=df.day,
             y=df.tip,
             name="tip",
             marker=dict(color="purple"))

data=[trace1,trace2]
layout=dict(title="BALIK",xaxis=dict(title="X"),yaxis=dict(title="Y"),barmode="relative")
fig=dict(data=data,layout=layout)
iplot(fig)
```

```python
dff=df.day.value_counts()
dff
```

# Pie Plot

```python
trace1=go.Pie(values=dff.values,labels=dff.index)
data=[trace1]
fig=dict(data=data)
iplot(fig)
```

```python
trace1=go.Pie(values=dff.values,labels=dff.index,hole=0.2)
data=[trace1]
fig=dict(data=data)
iplot(fig)
```

# Bubble Charts

```python
trace1=go.Scatter(x=df.tip,y=df.total_bill,marker=dict(size=20,showscale=True),mode="markers")
data=[trace1]
fig=dict(data=data)
iplot(fig)
```

# Histogram

```python
trace1=go.Histogram(x=dfsun.total_bill,marker=dict(color="pink",opacity=0.8),name="Sunday")
trace2=go.Histogram(x=dfsat.total_bill,marker=dict(color="yellow",opacity=0.8),name="Saturday")
trace3=go.Histogram(x=dfthur.total_bill,marker=dict(color="red",opacity=0.8),name="Thursday")
trace4=go.Histogram(x=dffri.total_bill,marker=dict(color="green",opacity=0.8),name="Friday")

data=[trace1,trace2,trace3]
layout=dict(barmode="overlay",title="DENEME",xaxis=dict(title="GNLER"))
fig=dict(data=data,layout=layout)
iplot(fig)
```

# BoxPlot

```python
trace1=go.Box(x=df.sex,y=df.total_bill)

data=[trace1]
fig=dict(data=data)
iplot(fig)
```

```python
trace1=go.Box(y=dfsat.total_bill,name="Saturday")
trace2=go.Box(y=dfsun.total_bill,name="Sunday")
trace3=go.Box(y=dfthur.total_bill,name="Thursday")
trace4=go.Box(y=dffri.total_bill,name="Friday")

data=[trace1,trace2,trace3,trace4]
fig=dict(data=data)
iplot(fig)
```

# 3D Scatter

```python
trace1=go.Scatter3d(x=df["size"],
                    y=df["tip"],
                    z=df["total_bill"],
                    mode="markers")
data=[trace1]
fig=dict(data=data)
iplot(fig)
```
