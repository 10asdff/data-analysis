```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
carona=pd.read_csv("data/carona.csv")
df=carona.copy()
```

```python
df.head()
```

```python
df.columns=["Sıra No","Tarih","Bölge/İl","Ülke","Son Güncelleme","Bağımlı","Ölümler","İyileşen"]
```

```python
df["Bölge/İl"].fillna("Unknown",inplace=True)
```

```python
df["Ülke"]=df["Ülke"].astype("category")
```

```python
df["Bölge/İl"]=df["Bölge/İl"].astype("category")
```

```python
ax=sns.barplot(x="Bağımlı",y="Ülke",data=df,ci=None,color="yellow",alpha=0.5)
ax=sns.barplot(x="Ölümler",y="Ülke",data=df,ci=None,color="blue",alpha=0.6)
ax=sns.barplot(x="İyileşen",y="Ülke",data=df,ci=None,color="red",alpha=0.5)
ax.set(xlabel='Ölümle/İyileşenler', ylabel='Ülkeler')
```
