# Olasılık Dağılımları

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
```

Olasılık : Olayın olabilirliğinin sayısal ifadesidir

### 1) Bernoulli Dağılımı

Başarılı (1) , başarısız (0) şeklinde iki sonuçlu olayda kullanılan olasılık dağılımıdır

Ör:

Mesela parayı 1 sefer attığımızda yazı gelme olasığı %50dir.(Yazı Gelme (Başarılı-1) - Yazı Gelmeme (Başarısız-0))

```python
yazi_olasilik=st.bernoulli(p=0.5)
aralik=np.linspace(0,1)
graf=sns.lineplot(aralik,yazi_olasilik.pmf(aralik))
graf.set(xlabel="YAZI GELMEME (0) - YAZI GELME (1)",ylabel="OLASILIK FREKANS")
```

```python
st.bernoulli(p=0.5).pmf(1)  # Yazı Gelmesinin Olasılık Değeri
```

```python
st.bernoulli(p=0.5).pmf(0) #Yazı Gelememesinin Olasılık Değeri
```

```python
st.bernoulli(p=0.5).pmf(0.5) #Diğer Olasılık Değeri
```

```python
yazi_olasilik=st.bernoulli.rvs(size=10000,p=0.5)
graf=sns.distplot(yazi_olasilik,kde=False,color="green")
graf.set(xlabel="YAZI GELMEME (0) - YAZI GELME (1)",ylabel="DAĞILIM FREKANS")
```

```python
yazi_olasilik[:10]
```

### 2) Binom Dağılımı

n deneme sonucu k başarılı olma olasılığı ile kullanılan olasılık dağılımıdır

Ör:

5 kere atılan zarın ön yüzünün 2 kez 3 gelmesinin olasılığı

```python
zar_olasilik=st.binom(n=5,p=1/6)
k=np.arange(10)
graf=sns.lineplot(k,zar_olasilik.pmf(k))
graf.set(xlabel="K Başarılı Olma Sayısı",ylabel="Olasılık Fonksiyonu")
```

```python
st.binom(5,1/6).pmf(2)
```

```python
zar_olasilik=st.binom.rvs(n=10,p=1/6,size=10000)
graf=sns.distplot(zar_olasilik,kde=False)
graf.set(xlabel="DENEME SAYISI")
graf.set(ylabel="DAĞILIM")
```

```python
len(zar_olasilik[zar_olasilik==5])/10000
```

```python
veriler=np.random.choice(zar_olasilik,size=100)
```

### 3) Poisson Dağılımı

Belirli bir zaman aralığında belirli bir alanda nadiren rastlanan olasılıkları hesaplamak için kullanılır

```python
n=4000
p=0.1
mu=4000*0.005
```

```python
mu
```

```python
k=np.arange(100)

ps=st.poisson(mu)
plt.plot(k,ps.pmf(k),"x-")
```

```python
veri_poission=st.poisson.rvs(mu=3,size=10000)

ax=sns.distplot(veri_poission,kde=False,color="green")
```
